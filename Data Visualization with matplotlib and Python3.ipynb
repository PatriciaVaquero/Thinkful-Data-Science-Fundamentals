<h1>Data Visualization with matplotlib and Python3</h1>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#make the plots big!
plt.rcParams['figure.figsize'] = (20.0, 10.0)
plt.rcParams.update({'font.size': 22})
```

```python
#import dataframe
df = pd.read_csv('cereal.csv')
```

<h2>Dataset = Nutrition Data on 80 Cereals</h2>
<h4>Source = https://www.kaggle.com/crawford/80-cereals</h4>
<h4>Content</h4>
<ul>Fields in the dataset:

<li>Name: Name of cereal
<li>mfr: Manufacturer of cereal
<li>A = American Home Food Products;
<li>G = General Mills
<li>K = Kelloggs
<li>N = Nabisco
<li>P = Post
<li>Q = Quaker Oats
<li>R = Ralston Purina
<li>type:
<li>cold
<li>hot
<li>calories: calories per serving
<li>protein: grams of protein
<li>fat: grams of fat
<li>sodium: milligrams of sodium
<li>fiber: grams of dietary fiber
<li>carbo: grams of complex carbohydrates
<li>sugars: grams of sugars
<li>potass: milligrams of potassium
<li>vitamins: vitamins and minerals - 0, 25, or 100, indicating the typical percentage of FDA recommended
<li>shelf: display shelf (1, 2, or 3, counting from the floor)
<li>weight: weight in ounces of one serving
<li>cups: number of cups in one serving
<li>rating: a rating of the cereals (Possibly from Consumer Reports?)</ul>

```python
df.calories.value_counts().plot(kind='bar')
plt.xlabel('Calories')
plt.ylabel('Number of Cereals')
plt.title('Average Calories per Serving of Cereals')
```

<p>A bar chart showing the average number of calories per serving of cereal.</p>
<p>But how big is a serving?</p>

```python
plt.scatter(df['calories'], df['cups'])
plt.xlabel('Calories')
plt.ylabel('Serving Size in Cups')
plt.title('Calories per Cup')
```

<p>A scatterplot to show the relationship between two variables.  This plot shows the relationship between calories and serving size</p>

```python
plt.hist(df['calories'])
plt.xlabel('Calories')
plt.ylabel('Number of Cereals')
plt.title('Average Calories per Serving of Cereals')
```

<p>A histogram is another way of showing the average calories per serving.  It looks like most cereals have between 90 and 130 calories per serving.</p>

```python
plt.subplot(1, 2, 1)
#subplot takes 3 arguments - the first 2 define the dimensions of the plot and the 3rd is the subplot you're making
plt.hist2d(df['cups'],df['calories'])
plt.xlabel('Cups')
plt.ylabel('Calories')
plt.title('Average Calories per Serving of Cereals')

plt.subplot(1, 2, 2)
plt.hist2d(df['cups'],df['protein'])
plt.xlabel('Cups')
plt.ylabel('Protein (grams)')
plt.title('Average Protein per Serving of Cereals')
```
